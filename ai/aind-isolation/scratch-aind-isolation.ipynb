```python
'''
Node:
- Helper class for creating trees
- A node contains a value and a list of sub-nodes
'''
class Node():
    
    def __init__(self, value):
        self._value = value
        self._sub_nodes = []

    def add_sub_node(self, sub_node):
        sub_node._parent = self
        self._sub_nodes.append(sub_node)
        return
    
    def create_sub_node(self, value):
        sub_node = Node(value)
        self._sub_nodes.append(sub_node)
        return


    @property
    def sub_nodes(self):
        return self._sub_nodes
    

    @property
    def value(self):
        return self._value

    @value.setter
    def value(self, v):
        self._value = v
    
    def is_leaf(self):
        return len (self._sub_nodes) == 0
    

    '''
    traverse_dfs_post_order:
    - does a tree traversal where root is self
    - traverses using a depth-first-search in post order
    - this is useful when calculating minimax score
    - fn_process_node is a function that is called on each node
    '''
    def traverse_dfs_post_order (self, fn_process_node):
        #--- does a dfs post-order traversal
        curr_node = self

        node_stack = []
        node_stack.append(curr_node)
        
        post_order_traversal = []
        
        #--- generate the post_order traversal list
        while node_stack:
            curr_node = node_stack.pop()
            
            post_order_traversal.append(curr_node)
            
            #--- we reverse the nodes to process the left most first
            node_stack.extend(curr_node.sub_nodes[::-1])

        #--- process each node in the post order traversal list
        #--- we start from the "right" of the list which is a leaf
        #--- the root will be the left-most (index 0)
        for node in post_order_traversal[::-1]:
            fn_process_node(node)
            
        return None
    

    '''
    methods below test functionality of the dfs_post order traversal
    '''
    def _fn_node_counter(self, node):
        self._size += 1
        return

    
    @property
    def size(self):
        self._size = 0
        self.traverse_dfs_post_order(self._fn_node_counter)
        return self._size
    
    
    
    def _fn_leaf_adder(self, node):
        
        if node.is_leaf():
            self._leaves.append(node)
        
        return Node
    
    
    @property
    def leaves(self):
        
        self._leaves = []
        
        self.traverse_dfs_post_order(self._fn_leaf_adder)
        
        return self._leaves

def fn_print_node_val(node):
    print (node.value)
        
n = [Node(n) for n in range(13)]
n[0].add_sub_node(n[1])
n[0].add_sub_node(n[2])
n[0].add_sub_node(n[3])

n[1].add_sub_node(n[4])
n[1].add_sub_node(n[5])
n[1].add_sub_node(n[6])

n[2].add_sub_node(n[7])
n[2].add_sub_node(n[8])
n[2].add_sub_node(n[9])

n[3].add_sub_node(n[10])
n[3].add_sub_node(n[11])
n[3].add_sub_node(n[12])

n[0].traverse_dfs_post_order(fn_print_node_val)

print (len(n[0].leaves))
```

```python
def create_minimax_game_tree(board, max_depth):
    """
    creates a game tree that can be searched using minimax.
    ---
    parameters:
        board: isolation.board from which to grow rest of tree 
            - the board is the current state of the board
            - the sub trees
        depth: depth of tree to make
            - depth of 0 returns current board state
            - depth of 1 is a game tree with all next legal moves
            - depth of 2 is a game tree with all next legal moves including depth 1's
            - etc...
        
    returns: a node that is the root of a game tree generated from board
    """
    root = Node(board)
    
    curr_node = root
    curr_depth = 0
    
    node_stack = []    #--- a stack is used to help generate the tree
    node_stack.append(curr_node)

    while node_stack:
        curr_node = node_stack.pop()

        #--- using the current board, generate boards from legal moves
        curr_board = curr_node.value

        #--- generate subtrees from all legal moves
        legal_moves = curr_board.get_legal_moves()

        for move in legal_moves:
            next_board = curr_board.forecast_move(move)    #--- forecast() automatically switches active player
            curr_node.add_sub_node(Node(next_board))        #--- next_board active player is inactive player of curr_board!
 
        curr_depth = curr_depth + 1
        if curr_depth < max_depth:
            #--- go deeper by adding the sub nodes to the stack
            node_stack.extend(curr_node.get_sub_nodes())
    
    return root
    
    
```

```python
class MinimaxNode(Node):
    '''
    A node where:
        - value[0] is the board
        - value[1] is the depth
        - vaule[2] is move that produced board
        - value[3] is minimax score of board
    '''
    
    def __init__ (self, board, depth=0, move=None, score=0):
        super().__init__([board, depth, move, score])

        return
        
    #--- board
    @property 
    def board(self):
        return self._value[0]
    
    @board.setter
    def board(self, board):
        self._value[0] = board
        
    #--- depth
    @property
    def depth(self):
        return self._value[1]
    
    @depth.setter
    def depth(self, depth):
        self._value[1] = depth
        
    #--- move
    @property
    def move(self):
        return self._value[2]
    
    @move.setter
    def move(self, move):
        self._value[2] = move
        
    #--- score
    @property
    def score(self):
        return self._value[3]
    
    @score.setter
    def score(self, score):
        self._value[3] = score
        
        
    def create_tree(self, max_depth):
        """
        creates a game tree that can be searched using minimax.
        ---
        parameters:
            board: isolation.board from which to grow rest of tree 
                - the board is the current state of the board
                - the sub trees
            depth: depth of tree to make
                - depth of 0 returns current board state
                - depth of 1 is a game tree with all next legal moves
                - depth of 2 is a game tree with all next legal moves including depth 1's
                - etc...

        returns: a node that is the root of a game tree generated from board
        """
        curr_minimax_node = self

        node_stack = []    #--- a stack is used to help generate the tree
        node_stack.append(curr_minimax_node)

        while node_stack:
            curr_minimax_node = node_stack.pop()

            #--- using the current board, generate boards from legal moves
            curr_board = curr_minimax_node.board

            #--- generate subtrees from all legal moves
            legal_moves = curr_board.get_legal_moves()
            next_depth = curr_minimax_node.depth + 1
            for move in legal_moves:

                #--- forecast() automatically switches active player
                next_board = curr_board.forecast_move(move)

                #--- next_board active player is inactive player of curr_board!
                new_sub_node = MinimaxNode(next_board, next_depth, move)

                curr_minimax_node.add_sub_node(new_sub_node)

            if next_depth < max_depth:
                #--- go deeper by adding the sub nodes to the stack
                node_stack.extend(curr_minimax_node.sub_nodes)

        return


    def _fn_calc_minimax_score(self, node):
        
        if node.is_leaf():
            #--- calculating leave score
            node.score = self._minimax_player.score(node.board, self._minimax_player)

        else:
            scores = [sub_node.score for sub_node in node.sub_nodes]
            print (node.depth, "- Branch sub node scores", scores)
            if node.board.active_player == self._minimax_player:
                #--- maximize 
                node.score = max(scores)
                print("choosing max score:", node.score)
                
            else:
                node.score = min(scores)
                print("choosing min score:", node.score)

            node._idx_best_sub_node = scores.index(node.score)
        
        return
    
    
    def best_move (self, minimax_player):
        
        if self.is_leaf():
            #--- no legal moves
            return (-1, -1)
        self._minimax_player = minimax_player
        self.traverse_dfs_post_order(self._fn_calc_minimax_score)
        print ("Best Move:", self._sub_nodes[self._idx_best_sub_node].move)
        print (self._sub_nodes[self._idx_best_sub_node].board.to_string())
        
        return self._sub_nodes[self._idx_best_sub_node].move
        
```
